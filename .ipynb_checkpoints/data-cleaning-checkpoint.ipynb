## Clean

#### Define

* Replace all records with the values, "Immediately", "As soon as possible", and "Immediate ..." in the StartDate column by "ASAP"
* Rename the nondescriptive column headers (ApplicationP, AboutC, RequiredQual, and JobRequirment) to descriptive column headers (ApplicationProcedure, AboutCompany, RequiredQualifications, and JobRequirements)

#### Code

```python
df_clean = df.copy()
```

* Rename the nondescriptive column headers (ApplicationP, AboutC, RequiredQual, and JobRequirment) to descriptive column headers (ApplicationProcedure, AboutCompany, RequiredQualifications, and JobRequirements)

```python
df_clean = df_clean.rename(columns={'ApplicationP': 'ApplicationProcedure',
                         'AboutC': 'AboutCompany',
                         'RequiredQual':'RequiredQualifications',
                         'JobRequirment':'JobRequirements'})
```

* Replace all records with the values, "Immediately", "As soon as possible", and "Immediate" in those cells in the StartDate column by the value "ASAP"

Find all the values that need to be replaced and put them in a list

```python
asap_list = ['Immediately', 'As soon as possible', 'Upon hiring',
             'Immediate', 'Immediate employment', 'As soon as possible.', 'Immediate job opportunity',
             '"Immediate employment, after passing the interview."',
             'ASAP preferred', 'Employment contract signature date',
             'Immediate employment opportunity', 'Immidiately', 'ASA',
             'Asap', '"The position is open immediately but has a flexible start date depending on the candidates earliest availability."',
             'Immediately upon agreement', '20 November 2014 or ASAP',
             'immediately', 'Immediatelly',
             '"Immediately upon selection or no later than November 15, 2009."',
             'Immediate job opening', 'Immediate hiring', 'Upon selection',
             'As soon as practical', 'Immadiate', 'As soon as posible',
             'Immediately with 2 months probation period',
             '12 November 2012 or ASAP', 'Immediate employment after passing the interview',
             'Immediately/ upon agreement', '01 September 2014 or ASAP',
             'Immediately or as per agreement', 'as soon as possible',
             'As soon as Possible', 'in the nearest future', 'immediate',
             '01 April 2014 or ASAP', 'Immidiatly', 'Urgent',
             'Immediate or earliest possible', 'Immediate hire',
             'Earliest  possible', 'ASAP with 3 months probation period.',
             'Immediate employment opportunity.', 'Immediate employment.',
             'Immidietly', 'Imminent', 'September 2014 or ASAP', 'Imediately']
```

```python
for old_value in asap_list:
    df_clean.StartDate.replace(old_value, 'ASAP', inplace=True)
```

#### Test

```python
# use df.info() to verify the change
df_clean.info()
```

```python
df_clean.StartDate.value_counts()
```

```python
# use the .value_counts, we can't be sure that everything was fixed
# we use a software testing technique -- assert statement
# .values is necessary, otherwise df_clean.StartDate's index will be checked instead of the values
for phrase in asap_list:
    assert phrase not in df_clean.StartDate.values
```

## Analysis & Visualization

```python
# Investigate the percentage of posting that are looking for someone to start urgently
# Number of 'ASAP' start dates (numerator)
asap_counts = df_clean.StartDate.value_counts()['ASAP']
asap_counts
```

```python
# Number of non-empty start dates (denominator)
non_empty_counts = df_clean.StartDate.count()
non_empty_counts
```

```python
# Percentage of positions with an urgent start date i.e. 'ASAP'
asap_counts / non_empty_counts
```

```python
%matplotlib inline
df_clean.StartDate.value_counts().plot(kind='pie');
```

```python
# Create a custom label and redraw the plot
import numpy as np
labels = np.full(len(df_clean.StartDate.value_counts()), "", dtype=object)
labels[0] = 'ASAP'
df_clean.StartDate.value_counts().plot(kind='pie', labels=labels);
```
